```python
prodigal_dir = "/home/vini/data/prosyn/processing/prodigal/"
genes_dir = path.join(prodigal_dir, "genes")
proteins_dir = path.join(prodigal_dir, "proteins")
cds_dir = path.join(prodigal_dir, "cds")
```

```python
def prodigal_files(Genome):
    Genome.files["prodigal"] = dict()
    for k, v, x in zip(("genes", "proteins", "cds"), (genes_dir, proteins_dir, cds_dir), (".fna", ".faa", ".gbk")):
        Genome.files["prodigal"][k] = path.join(v, Genome.name + f"_prodigal_{k}{x}" )
```

```python
for genome in df["abacat_file"]:
    prodigal_files(genome)
    for k, v in genome.files["prodigal"].items():
        if not path.isfile(v):
            print(genome.name, v)
```

```python
output_dir = "/home/vini/data/prosyn/processing/dmnd_cog/"

for genome in df["abacat_file"]:
    genome.files["dmnd_out"] = path.join(output_dir, genome.name + "_dmnd.out")
```

```python
def run_diamond(genome, db):
    query = genome.files["prodigal"]["proteins"]
    output = genome.files["dmnd_out"]
    cmd = f"diamond blastp -q {query} \
            --db {db} -o {output} -f 6 \
            qseqid sseqid pident length \
            mismatch gapopen qstart qend \
            sstart send evalue bitscore score \
            -p 60 -k 1"
    system(cmd)
    print(f"Processed {genome.name}.")
```

```python
cog_dir = path.dirname(config.CONFIG["db"]["COG"])
cog_domains = pd.read_csv(
    path.join(cog_dir, "cog2003-2014.csv")).iloc[:, :-1]
cog_domains.columns = (
    "domain-id", "genome-name",
    "protein-id","protein-length",
    "domain-start", "domain-end",
    "COG-id", "membership-class"
)
cog_domains.drop_duplicates(subset="domain-id", inplace=True)
cog_domains.set_index("domain-id", inplace=True)
cogt = COGTranslator()
```

```python
def get_COG_name(sseqid):
    """
    Provide a sseqid from diamond output and returns
    (COG-XXXX, genome_name, description, metabolism)
    """
    domain = int(sseqid.split("|")[1])
    row = cog_domains.loc[domain]
    COG_code = row["COG-id"]
    genome_name = row["genome-name"]
    description = cogt.name_from_code(COG_code)
    # metabolism = cogt.letter_from_code(COG_code)
    
    return (sseqid, COG_code, genome_name, description)#, metabolism)
```

```python
def generate_dmnd_df(dmnd_out):
    dmnd = pd.read_csv(dmnd_out, sep="\t")
    dmnd.columns = "qseqid sseqid pident length \
                mismatch gapopen qstart qend \
                sstart send evalue bitscore score".split()
    dmnd = dmnd.drop_duplicates(subset=("qseqid"))
    COG_names = dmnd["sseqid"].apply(lambda s: get_COG_name(s))
    COG_names = pd.DataFrame(COG_names.tolist(), columns=("sseqid COG_code genome_name description".split()))
    dmnd = dmnd.merge(COG_names, on="sseqid")
    return dmnd
```
